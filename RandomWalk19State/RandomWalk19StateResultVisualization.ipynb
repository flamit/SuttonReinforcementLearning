```python
from RandomWalk19State import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
td_alpha_array = np.arange(0, 1.02, 0.02)
num_step_look_ahead_array = 2**np.arange(0,9)

rms_error_td_dict = np.zeros((len(num_step_look_ahead_array),len(td_alpha_array)))

for step_index, step in enumerate(num_step_look_ahead_array):
    for alpha_index, alpha in enumerate(td_alpha_array):
        rms_error_td_dict[step_index][alpha_index] = get_rms_error(num_episode=10, 
                                                                   alpha=alpha, 
                                                                   num_step_look_ahead=step,
                                                                   num_instance=500)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
fontsize = 15
for index, step in enumerate(num_step_look_ahead_array):
    ax.plot(td_alpha_array, 
            rms_error_td_dict[index],
            linewidth=2,
            linestyle='-',
            label='TD step='+str(step))
ax.set_xlim([min(td_alpha_array),max(td_alpha_array)])
ax.grid(linestyle='--')
ax.legend(loc='best', fontsize=fontsize)
ax.set_xlabel('alpha', fontsize=fontsize)
ax.set_ylabel('RMS error', fontsize=fontsize)
ax.set_title('RMS error averaged over states and 100 realizations', fontsize=fontsize)
ax.tick_params(labelsize=fontsize)
plt.show()
```
